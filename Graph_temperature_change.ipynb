# **Important Libraries**

```python
# Libraries for Data Manipulation
import pandas as pd
# Libraries for Visualization
import plotly.express as px
```

# **Loading and Exploring the Data**

```python
temp = pd.read_csv("temperature_change_data_11-29-2021.csv")
temp.head()
```

```python
# Data Summary
temp.describe()
```

# **Preparing the Data**

```python
# Grouping the necessary data
temp = temp.groupby(["Area", "Area Code (ISO3)", "Year"])["Value"].sum().reset_index()
```

# **Graph of Temperature Change in whole world**

```python
# Visualizing the Data
px.choropleth(temp, 
              locations = 'Area Code (ISO3)',
              color="Value", 
              animation_frame="Year",
              color_continuous_scale="Oranges",             
              scope="world",
              title='<b>Temperature Change in the World<br><sup>1961 to 2020',
              range_color=(-5.979, 8.206),
              height=600
             )  
```

#  **Graph of Temperature Change in South Amercia**

```python
# Visualizing the Data
px.choropleth(temp, 
              locations = 'Area Code (ISO3)',
              color="Value", 
              animation_frame="Year",
              color_continuous_scale="ylorbr",             
              scope="south america",
              title='<b>Temperature °C Change in the South America<br><sup>1961 to 2020',
              range_color=(-5.979, 8.206),
              height=600
             )  
```

# **Seasonal Temperature Change in the World**

```python
# preparing data according to the seasonal temperature change 
temp1= pd.read_csv("temperature_change_data_11-29-2021.csv")
evolution = temp1[ (temp1["Months"] != "Meteorological year") ].reset_index(drop=True)

evolution = evolution.groupby(["Area", "Area Code (ISO3)", "Months"])["Value"].sum().reset_index()
evolution.head()
```

```python
# Visualizing the Data
px.choropleth(evolution, 
              locations = 'Area Code (ISO3)',
              color="Value", 
              animation_frame="Months",
              color_continuous_scale="redor",             
              scope="world",
              title='<b>Seasonal Temperature °C Change in World<br><sup>1961 to 2020',
              height=600
             )  
```
